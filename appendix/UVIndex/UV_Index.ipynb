```python
# Import all dependencies
import requests
import json
import pandas as pd
import wikipedia as wp
import pyowm
from config import api_key
```

```python
# Load city_df3 CSV file into dataFrame
df = pd.read_csv('california_cities.csv')
df.head(10)
 
```

```python
# City, Latitude, and Longitude
city_df = df.drop(['Unnamed: 0'], axis=1)
city_df.head()
```

```python
# import the module OWM
owm = pyowm.OWM(api_key)
```

```python
# Assign the coordinates to uvi
uvi = owm.uvindex_around_coords(40.712784, -74.005941)
```

```python
# A forecast UV Index for New York City
uv_list = owm.uvindex_forecast_around_coords(40.71, -74.00)
uv_list[0]
```

```python
# the value of the UV Index for the assigned city fro, UVI
uvi.get_value()
```

```python
# Loop through the dataframe and push the UV Index for each city to uv_index
uv_index = []
for i, r in city_df.iterrows():
    uvi = owm.uvindex_around_coords(r.loc['lat'], r.loc['lng'])
    uv_index.append(uvi.get_value())
```

```python
# Make the list into series
uv = pd.Series(uv_index)
```

```python
# Add the values to city_df
city_df['UV_value'] = uv.values
```
