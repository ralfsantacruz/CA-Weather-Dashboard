```python
from sqlalchemy import create_engine
from graph_gen import get_coords, pollution
from pprint import pprint
import pandas as pd
import json
import requests
```

```python
engine = create_engine("mysql://root:password@localhost/AQI")
```

```python
x,y = get_coords('san francisco')
```

```python
p = pollution(x,y)
```

```python
pprint(p)
```

```python
def consolidate_aqi(data):
    dic = {
        'datetime' : [],
        'aqi': [],
        'aqi_display': [],
        'category': [],
        'color': [],
        'display_name': [],
        'dominant_pollutant': []
          }
    
    for entry in data['data']:
        dic['datetime'].append(entry['datetime'])
        dic['aqi'].append(entry['indexes']['baqi']['aqi'])
        dic['aqi_display'].append(entry['indexes']['baqi']['aqi_display'])
        dic['category'].append(entry['indexes']['baqi']['category'])
        dic['color'].append(entry['indexes']['baqi']['color'])
        dic['display_name'].append(entry['indexes']['baqi']['display_name'])
        dic['dominant_pollutant'].append(entry['indexes']['baqi']['dominant_pollutant'])
    
    return pd.DataFrame(dic)
```

```python
df = consolidate_aqi(p)
```

```python
df.to_sql(name="san_francisco",con=engine)
```
