```python
# from config import owm_api_key
import requests

import pandas as pd
from pprint import pprint

import time
```

```python
df = pd.read_csv('cali_cities.csv').drop(columns='Unnamed: 0')
df.head()
```

```python
owm_api_key = "81865b5f49cc2527ef3402a21ce511a8"
```

```python
rain = []

for lat,lng in zip(df['lat'][400:410],df['lng'][400:410]):
    base_url = f"http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lng}&APPID={owm_api_key}"
    data = requests.get(base_url).json()
    pprint(data)
    
    try:
        rain.append(data['rain']['1h'])
    except:
        rain.append(0)
    
#     time.sleep(1.5)
    
    
print(rain)
```
